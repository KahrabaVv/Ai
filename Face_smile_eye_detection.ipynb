```python
import cv2
from matplotlib import pyplot as plt
img=cv2.VideoCapture(0)
i=0
while img.isOpened():
      stat, frame=img.read()
      if stat ==True:
                 
            k=cv2.waitKey(1)
            if k==ord('q'):
                  break
            if k==ord('s'):
                  cv2.imwrite('img'+str(i)+'.png',frame)
                  i+=1
            gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
            font=cv2.FONT_HERSHEY_SIMPLEX
            faceClassifier=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
            faces=faceClassifier.detectMultiScale(gray,scaleFactor=1.09,minNeighbors=5,minSize=(80,80))
            eyeClassifier=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_eye.xml')
            smileclassifier=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_smile.xml')
            
            for x,y,w,h in faces: # comment x,y,w,h
                  cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
                  face_slice=gray[y:y+h,x:x+w]
                  eyes=eyeClassifier.detectMultiScale(face_slice,scaleFactor=1.09,minNeighbors=6,minSize=(12,12))
                  smiles=smileclassifier.detectMultiScale(face_slice,scaleFactor=1.1,minNeighbors=6,minSize=(60,60))
                  cv2.putText(frame,"me",(x+33,y+h+50),font,1,(255,255,0),2)
                  for x3,y3,w3,h3 in smiles:
                        cv2.rectangle(frame,(x+x3,y+y3),(x+x3+w3,y+y3+h3),(100,150,500),5)
                        cv2.putText(frame,"smile",(x3+250,y3+170),font,1,(255,255,0),2)
                  for x2,y2,w2,h2 in eyes:
                        x01=int( (x+x2+x+x2+w2)/2)
                        x02=int ((y+y2+h2+y+y2)/2)
                        z= int(w2/2)
                        cv2.circle(frame,(x01,x02),z,(255,0,100),5)
                        cv2.putText(frame,"eye",(x+x2+w2-40,y+y2+h2+20),font,1,(255,255,0),2)
                        
            cv2.imshow("my_video",frame) 
            
            
      else:
            k=cv2.waitKey(1)
            if k==ord('q'):
                  break
            cv2.destroyAllWindows()
            print("wrong frame") 
            break  


img.release()
cv2.destroyAllWindows()   
```
